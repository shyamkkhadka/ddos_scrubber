```python
# Normalize IPV4 normalized by /24.
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
from ipaddress import ip_network

year = "2022"
scrubber = "32787"

# Data are in customers_prefixes_scrubber_32787_2024.csv
df = pd.read_csv("../data/customers_prefixes_scrubber_"+scrubber+"_"+year+".csv")
prefix_list = df["prefix"]

def normalize_prefixes(prefixes, output_file="../data/normalized_prefixes_"+scrubber+"_"+year+".txt"):
    normalized = set()  # Use a set to avoid duplicates
    for prefix in prefixes:
        network = ip_network(prefix, strict=False)

        if network.version == 4:
            if network.prefixlen < 24:
                normalized.update(map(str, network.subnets(new_prefix=24)))  # Split smaller subnets
            else:
                normalized.add(str(network))
        elif network.version == 6:
            normalized.add(str(ip_network(f"{network.network_address}/64", strict=False)))  # Normalize IPv6
    
    print("Total prefixes are %s in year %s" %(len(normalized),year))

    # Write to file in bulk
#     with open(output_file, "w") as f:
#         f.write("\n".join(sorted(normalized)) + "\n")
        
# # Example usage
# prefix_list = [
#     "192.168.0.0/23",  # Splits into two /24s
#     "10.0.0.0/20",  # Splits into 16 /24s
#     "2001:db8::/48",
#     "172.16.5.0/25",
# ] 

normalize_prefixes(prefix_list)

```
